```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
from sklearn.model_selection import StratifiedKFold, KFold

# import gensim.parsing.preprocessing as preprocessing
```

```python
arr = []
def preprocess(dataframe):
  for n, row in dataframe.iterrows():
    #preprocess here
    review = row["body"]
    # print(review)
    arr.append(review)
  return arr


df = pd.read_csv("/content/drive/MyDrive/CSC_522_ALDA/balanced_data_set_3000.csv").dropna()
df['rating'] = df['rating'].astype(float)

a = preprocess(df)
print(len(a))

del(df["body"])

df["body"] = a
```

```python
df.columns
```

```python
var = 0

X_train=pd.DataFrame(df.sample(frac=0.8,random_state=200)[['body','generalized_ai_rating']])
X_test=pd.DataFrame(df.drop(X_train.index)[['body','generalized_ai_rating']])

print(X_train['generalized_ai_rating'].value_counts())
print(X_test['generalized_ai_rating'].value_counts())
```

```python
model = ClassificationModel(
    'bert', 
    'bert-base-cased', 
    num_labels=3, 
    use_cuda=False,
    args={'num_train_epochs':1, 'reprocess_input_data': True, 'overwrite_output_dir': True}
)
#works
```

```python
model.train_model(X_train, overwrite_output_dir=True)
```

```python
result, model_outputs, wrong_predictions = model.eval_model(X_test[['body','generalized_ai_rating']])
```

```python
y_pred = []
for val in model_outputs:
  y_pred.append(np.where(val == max(val))[0])
```
