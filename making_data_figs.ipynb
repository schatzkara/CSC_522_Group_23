```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
all = pd.read_csv("/content/drive/MyDrive/combined.csv").dropna()
all.head()
```

```python
product_counts = all["product"].value_counts()
# print(product_counts)
```

```python
colors={'boxes': 'black', 'whiskers': 'black', 'fliers': 'black', 'means': 'black', 'medians': 'gray', 'caps': 'black'} 
linewidths={'boxes': 2, 'whiskers': 2, 'fliers': 2, 'means': 2, 'medians': 2, 'caps': 2}

plt.figure(1, (6, 1))  # something, plot size
p = plt.subplot(111)

bp = plt.boxplot(product_counts, labels=[""], showfliers=True, vert=False, flierprops=dict(marker='o', markersize=2, linestyle='none'), widths=0.25, positions=[0])
for element in ['boxes', 'whiskers', 'fliers', 'means', 'medians', 'caps']:
  plt.setp(bp[element], color=colors[element], linewidth=linewidths[element])

plt.title("")
# plt.xlabel("Number of Reviews Per Product", fontsize=10)
# plt.ylabel("", fontsize=10)
p.set_ylim(-0.3, 0.3)
```

```python
plt.figure(1, (5, 3))  # something, plot size
p = plt.subplot(111)

bp = plt.hist(product_counts)

plt.title("")
plt.xlabel("Number of Reviews Per Product", fontsize=10)
plt.ylabel("", fontsize=10)
```
