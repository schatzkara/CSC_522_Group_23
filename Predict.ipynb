```python
from simpletransformers.classification import ClassificationModel

model = ClassificationModel("xlnet", "/content/drive/MyDrive/CSC522/xlnet/5_epoch_xlnet/checkpoint-4500-epoch-5", use_cuda=False)
```

```python
model_trained_data = pd.read_csv("/content/drive/MyDrive/CSC522/data/balanced_data_set_3000.csv", index_col = 0).dropna()
```

```python
all = pd.read_csv('/content/drive/MyDrive/CSC522/data/combined.csv', index_col=0).drop(model_trained_data.index).sample(5000, random_state=100)
```

```python
all
```

```python
revs = all['body'].to_list()
```

```python
predictions, raw_outputs = model.predict(revs)
```

```python
raw = raw_outputs.tolist()
```

```python
preds = np.array(predictions) - 1
```

```python
all["xlnet_prediction"] = preds
```

```python
all["raw_output_score"] = raw
```

```python
all.to_csv("/content/drive/MyDrive/CSC522/data/xlnet_preds.csv")
```
