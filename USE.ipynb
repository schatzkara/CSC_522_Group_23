```python
from sklearn.model_selection import train_test_split

X = all["body"]
y = all["generalized_rating"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=200)
```

```python
import gensim.parsing.preprocessing as preprocessing
import sklearn.metrics
import pprint as pp
import numpy as np
import gensim
import nltk

from gensim.models import Word2Vec
from sklearn.neighbors import KNeighborsClassifier

nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
```

```python
#preprocess function
def preprocess(data):
  lines = []
  for r in data:
    try:
      # preprocess here
      lines.append(r)
    except:
      lines.append(r)
  return lines
```

```python
# Load tensor flow universal sentence encoder model
module_url = "https://tfhub.dev/google/universal-sentence-encoder/4"
model = hub.load(module_url)
print ("module %s loaded" % module_url)
def embed(input):
  return model(input)
```

```python
# Preprocess data
messages = preprocess(X_train)
messages_test = preprocess(X_test)


messages_labels = []
messages_test_labels = []

# Reduce logging output.
logging.set_verbosity(logging.ERROR)

# create embeddings for the data
def create_embeddings(messages_list):
  return np.array(embed(messages_list))

# create x train and test
x_train = create_embeddings(messages)
x_test = create_embeddings(messages_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

def analyze_resuts(x_train, y_train, x_test, y_test):
    # clf = KNeighborsClassifier(n_neighbors=4)
    # clf = DecisionTreeClassifier(random_state=10)
    # clf = RandomForestClassifier(random_state=10)
    clf = AdaBoostClassifier(random_state=10)

    clf.fit(x_train, y_train)

    acc1 = clf.score(x_train, y_train)
    acc2 = clf.score(x_test, y_test)

    print("Baseline accuracy: " + str(acc1))
    print("Tested accuracy: " + str(acc2))

    y_pred = clf.predict(x_test)

    labels = [-1,0,1]

    report = sklearn.metrics.classification_report(
        y_test, y_pred, digits=3, labels=labels
    )

    print("Classification report:")
    print(report)

    cm = sklearn.metrics.confusion_matrix(y_test, y_pred, labels=labels)
```

```python
analyze_resuts(x_train, y_train, x_test, y_test)
```
