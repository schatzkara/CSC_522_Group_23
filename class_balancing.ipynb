```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/CSC522/data/combined_generalized_rating.csv", index_col=0).dropna()
df.head()
```

```python
df["generalized_rating"].value_counts()
```

```python
sample_df = df.groupby("generalized_rating").sample(n=3000, random_state=1, replace=True)
sample_df
```

```python
sample_df["generalized_rating"].value_counts()
```

```python
generalized_ai_rating = []
for val in sample_df["generalized_rating"]:
  if val == -1:
    generalized_ai_rating.append(0)
  if val == 0:
    generalized_ai_rating.append(1)
  if val == 1:
    generalized_ai_rating.append(2)
  
sample_df["generalized_ai_rating"] = generalized_ai_rating
sample_df
```

```python
sample_df["generalized_ai_rating"].value_counts()
```

```python
sample_df.to_csv("/content/drive/MyDrive/CSC522/data/balanced_data_set_3000.csv")
```
