```python
import gensim.parsing.preprocessing as preprocessing
import sklearn.metrics
import pprint as pp
import numpy as np
import gensim
import nltk

from gensim.models import Word2Vec
from sklearn.neighbors import KNeighborsClassifier

nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
```

```python
#preprocess function
def preprocess(data):
  lines = []
  for r in data:
    try:
      # preprocess here
      lines.append(r)
    except:
      lines.append(r)
  return lines

# tokenize data per sentence using nltk
def tokenize(data):
  tokenized = []
  for sentence in data:
    try:
      tokenized.append (nltk.word_tokenize(sentence))
    except:
      continue;
  return tokenized
```

```python
x_train = preprocess(X_train)
x_test = preprocess(X_test)
```

```python
# tokenize raw all data and preprocessed all data
tokenized_sentences_1 = tokenize(X_train)
tokenized_sentences_2 = tokenize(X_test)
tokenized_sentences_3 = tokenize(x_train)
tokenized_sentences_4 = tokenize(x_test)
# tokenized_sentences = tokenized_sentences_1 + tokenized_sentences_2 <- uncomment when there is preprocessing data
tokenized_sentences = tokenized_sentences_1 + tokenized_sentences_2 + tokenized_sentences_3 + tokenized_sentences_4

# create a Word2Vec model using tokenized values and CBOW
model = Word2Vec(tokenized_sentences, min_count=1) #CBOW
print(model)
```

```python
path = "/content/drive/MyDrive/CSC522/"
learned_words = list(model.wv.index_to_key)
model.wv['case']
model.save(path + "word2Vec.model")
```

```python
model_loaded = Word2Vec.load(path + "word2Vec.model")
```

```python
model_loaded.wv['try']
```

```python
# create embedding vectors using model
def create_vectors(data):
  sentence_embedding = []
  for sentence in data:
    tokens = sentence.split(' ')

    word_embedding = []
    for word in tokens:
      word_embedding.append(model_loaded.wv['word'])
    m = np.vstack(word_embedding)
    v = np.mean(m, axis=0)

    sentence_embedding.append(v)
  return sentence_embedding

# create matricies for train and test data
x_train_embedding = create_vectors(x_train)
x_test_embedding = create_vectors(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

def analyze_resuts(x_train, y_train, x_test, y_test):
    # clf = KNeighborsClassifier(n_neighbors=4)
    # clf = DecisionTreeClassifier(random_state=10)
    # clf = RandomForestClassifier(random_state=10)
    clf = AdaBoostClassifier(random_state=10)

    clf.fit(x_train, y_train)

    acc1 = clf.score(x_train, y_train)
    acc2 = clf.score(x_test, y_test)

    print("Baseline accuracy: " + str(acc1))
    print("Tested accuracy: " + str(acc2))

    y_pred = clf.predict(x_test)

    labels = [-1,0,1]

    report = sklearn.metrics.classification_report(
        y_test, y_pred, digits=3, labels=labels
    )

    print("Classification report:")
    print(report)

    cm = sklearn.metrics.confusion_matrix(y_test, y_pred, labels=labels)
```

```python
analyze_resuts(x_train_embedding, y_train, x_test_embedding, y_test)
```
